```python
import numpy as np 
tensor_1 = np.arange(1,11)
print("1D Tensor: " , tensor_1)
print("Sum: " , np.sum(tensor_1) )
print("Multiplied: " , (tensor_1 * 2))
```

# 2D Tensor 

```python
array1 = np.array([[1,2,3] , [3,4,5] , [6,7,8]])
array1
```

```python
array1.ndim
```

```python
import numpy as np 
tensor_2 = np.arange(1,10).reshape(3,3)
print("Tensor 2D: " , tensor_2)
print("sum of each row: " , np.sum(tensor_2 , axis=1))
print("sum of each coulmn: " , np.sum(tensor_2 , axis=0))
print("Multiplied by 3:\n", tensor_2 * 3)
```

# 3D Tensors

```python
tensor_3d = np.arange(1, 25).reshape(2, 3, 4)  # 2x3x4 tensor
print("3D Tensor:\n", tensor_3d)
print("Sum along last axis (columns):\n", np.sum(tensor_3d, axis=2))
print("Maximum value:", np.max(tensor_3d))
print("\n")
```
