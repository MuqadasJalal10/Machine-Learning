## 1 - Scatterplot(Numerical - Numerical)

```python
sns.scatterplot(x='total_bill' , y='tip' , data=tip)
```

```python
sns.scatterplot(x='total_bill' , y='tip' , data=tip , hue=tip['sex'])
```

```python
sns.scatterplot(x='total_bill' , y='tip' , data=tip , hue=tip['sex'] , style=tip['smoker'])
```

## Barplot ( Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(x='Pclass' , y='Age' , data=titanic , hue=titanic['Sex'])
```

## Boxplot (Numerical - Categorical)

```python
sns.boxplot(x = 'Pclass' , y = 'Age' , data=titanic)
```

## Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived'] ==0] ['Age'] , hist=False)
sns.distplot(titanic[titanic['Survived'] ==1] ['Age'] , hist=False)
```

## Heatmap (Categorical - Categorical)

```python
titanic.head()
```

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```
